```python
cols = []
for i in range(29):
    cols.append(df[i][0])
print(cols)
len(cols)
```

```python
df.columns = cols
```

```python
df.index
```

```python
df = df.drop(["Year"],axis=0)
```

```python
df.columns
```

```python
df.dtypes
```

```python
df = df.astype(float)
```

```python
meean = df['Arghakhanchi (Khanchikot)'].mean()
for place in cols:
    df[place] = df[place].fillna(df[place].mean())
df
```

```python
df = df.append(df.sum().rename('Total'))
df
```

```python
trans = df.T
trans
```
